```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importinig the dataset train
train = pd.read_csv('/content/drive/MyDrive/Dengue/dengue_features_train.csv')
```

```python
train
```

```python
#Importinig the dengue_labels
lables = pd.read_csv('/content/drive/MyDrive/Dengue/dengue_labels_train.csv')
lables
```

```python
#Take the total_cases from dengue_labes to dataset train because the total_cases column is the target column of this project
train['total_cases'] = lables['total_cases']
```

```python
train
```

```python
#Starting the pre_processing with see the NaN values
train.isnull().sum()
```

```python
#Transforming the categorical column for the number column
from sklearn.preprocessing import LabelEncoder
lb= LabelEncoder()
train['city'] = lb.fit_transform(train['city'])
```

```python
#Changing the columns date
#Spliting the column weel_start_date in column Month and day the year column already exist
import datetime as dt
train['week_start_date'] = pd.to_datetime(train['week_start_date'])
train['Month'] = train['week_start_date'].dt.month
train['day'] = train['week_start_date'].dt.day
```

```python
#Drop the week_start_date because don't is necessary
train.drop('week_start_date', axis=1, inplace=True)
```

```python
#switching the NaN values for mean each column
for i in train.columns:
  train[i].fillna(train[i].mean(), inplace=True)
```

```python
#Spliting the train dataset for train_model and test_model
train_model = train.drop('total_cases', axis=1)
test_model = train['total_cases']
```

```python
#See and treating the outliers values
from pyod.models.knn import KNN
detector = KNN()
detector.fit(train_model)
```

```python
predict = detector.labels_
len(predict)
```

```python
#We have 1310 values not outliers(0) and 146 outliers(1)
np.unique(predict, return_counts=True)
```

```python
#See the distance calculation for outliers detection
trust = detector.decision_scores_
trust
```

```python
#Doing a list just outliers indexs
outliers = []
for i in range(len(predict)):
  if predict[i] == 1:
    outliers.append(i)
```

```python
#Removing the outliers values for mean of each column
column = []
for i in train.columns:
  column.append(i)
for c in column:
  for o in outliers:
    train.loc[o,c] = train.loc[:,c].mean()
```

```python
#Normalixation the dataset
from sklearn.preprocessing import MinMaxScaler
obj_norm = MinMaxScaler().fit(train_model)
train_norm = obj_norm.transform(train_model)
train_norm
```

```python
train_model.columns
```

```python
train_norm = pd.DataFrame(train_norm)
train_norm.columns = ['city', 'year', 'weekofyear', 'ndvi_ne', 'ndvi_nw', 'ndvi_se',
       'ndvi_sw', 'precipitation_amt_mm', 'reanalysis_air_temp_k',
       'reanalysis_avg_temp_k', 'reanalysis_dew_point_temp_k',
       'reanalysis_max_air_temp_k', 'reanalysis_min_air_temp_k',
       'reanalysis_precip_amt_kg_per_m2',
       'reanalysis_relative_humidity_percent', 'reanalysis_sat_precip_amt_mm',
       'reanalysis_specific_humidity_g_per_kg', 'reanalysis_tdtr_k',
       'station_avg_temp_c', 'station_diur_temp_rng_c', 'station_max_temp_c',
       'station_min_temp_c', 'station_precip_mm', 'Month', 'day']
```

```python
train_norm
```

```python
#standardizing values
from sklearn.preprocessing import StandardScaler
obg_df = StandardScaler().fit(train_model)
train_df = obg_df.transform(train_model)
train_df
```

```python
train_df = pd.DataFrame(train_df)
train_df.columns = ['city', 'year', 'weekofyear', 'ndvi_ne', 'ndvi_nw', 'ndvi_se',
       'ndvi_sw', 'precipitation_amt_mm', 'reanalysis_air_temp_k',
       'reanalysis_avg_temp_k', 'reanalysis_dew_point_temp_k',
       'reanalysis_max_air_temp_k', 'reanalysis_min_air_temp_k',
       'reanalysis_precip_amt_kg_per_m2',
       'reanalysis_relative_humidity_percent', 'reanalysis_sat_precip_amt_mm',
       'reanalysis_specific_humidity_g_per_kg', 'reanalysis_tdtr_k',
       'station_avg_temp_c', 'station_diur_temp_rng_c', 'station_max_temp_c',
       'station_min_temp_c', 'station_precip_mm', 'Month', 'day']
```

```python
train_df
```

```python
train_norm
```

```python
#Selecting the best columns for learning of model
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
train_k_best = SelectKBest(chi2, k=15).fit_transform(train_norm, test_model)
```

```python
train_norm
```

```python
train_k_best = pd.DataFrame(train_k_best)
```

```python
train_k_best
train_k_best.columns = ['city', 'year', 'weekofyear','ndvi_se', 'ndvi_sw', 'precipitation_amt_mm', 'reanalysis_air_temp_k', 'reanalysis_max_air_temp_k',
                        'reanalysis_min_air_temp_k', 'reanalysis_sat_precip_amt_mm', 'reanalysis_tdtr_k', 'station_diur_temp_rng_c',
                        'station_min_temp_c', 'Month', 'day']
```

```python
train_k_best
```

```python
train_norm.columns
```

```python
#Make the division in x_train, x_test , y_train, y_test to train the model and test_size 0.15 (based in experiments the 0.15 value is the best)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_k_best, test_model, test_size = 0.15, random_state = 4)
```

```python
#We're use the XGBRegressor because this model is the most acertive
#We have somes paramns we got a best score but for don't have overfitting some standard parameters were maintained
from math import gamma
import xgboost
xgb = xgboost.XGBRegressor(max_depth=6, eta = 0.4, gamma = 1)
xgb = xgb.fit(x_train, y_train)
```

```python
#score: 86.26
ac = xgb.score(x_test, y_test)
print('Accurancy :' + str(ac))
```

```python
x_pred = xgb.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, x_pred)
```
